## RFM Analysis

menghitung RFM value

menggabungkan tabel `orders` dan `order_items`

```python
orders_items_df = pd.merge(orders_df, order_items_df, on='order_id')
```

menghitung total nilai setiap pesanan

```python
orders_items_df['total_price'] = orders_items_df['price'] + orders_items_df['freight_value']
```

menghitung RFM values

```python
rfm_df = orders_items_df.groupby('customer_id').agg({
    'order_purchase_timestamp': lambda x: (orders_df['order_purchase_timestamp'].max() - x.max()).days,  # Recency
    'order_id': 'count',  # Frequency
    'total_price': 'sum'  # Monetary Value
})
rfm_df.rename(columns={
    'order_purchase_timestamp': 'Recency',
    'order_id': 'Frequency',
    'total_price': 'MonetaryValue'
}, inplace=True)
```

menentukan RFM score

```python
rfm_df['R_score'] = pd.qcut(rfm_df['Recency'], q=4, labels=range(4, 0, -1))
rfm_df['F_score'] = pd.qcut(rfm_df['Frequency'].rank(method='first'), q=4, labels=range(1, 5))
rfm_df['M_score'] = pd.qcut(rfm_df['MonetaryValue'], q=4, labels=range(1, 5))
```

menggabungkan RFM score

```python
rfm_df['RFM_Score'] = rfm_df['R_score'].astype(str) + rfm_df['F_score'].astype(str) + rfm_df['M_score'].astype(str)
```

segmentasi pelanggan berdasarkan RFM score

```python
def segment_customer(rfm_score):
    if rfm_score >= '444':
        return 'Champions'
    elif rfm_score >= '333':
        return 'Loyal Customers'
    elif rfm_score >= '222':
        return 'Potential Loyalists'
    elif rfm_score >= '111':
        return 'New Customers'
    else:
        return 'Need Attention'

rfm_df['Customer_Segment'] = rfm_df['RFM_Score'].apply(segment_customer)
```

### Clustering

mengelompokkan pelanggan menggunakan KMeans dengan jumlah claster 3

```python
from sklearn.cluster import KMeans

n_clusters = 3

# Membuat model K-Means
kmeans = KMeans(n_clusters=n_clusters, random_state=0)

# Melakukan clustering pada RFM values
rfm_df['Cluster'] = kmeans.fit_predict(rfm_df[['Recency', 'Frequency', 'MonetaryValue']])

print(rfm_df.head())

plt.figure(figsize=(10, 6))
plt.scatter(rfm_df['Recency'], rfm_df['Frequency'], c=rfm_df['Cluster'], cmap='viridis')
plt.title('Clustering Pelanggan Berdasarkan RFM Values')
plt.xlabel('Recency')
plt.ylabel('Frequency')
plt.show()
```
